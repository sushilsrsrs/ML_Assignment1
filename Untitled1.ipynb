```python
import pandas as pd
df=pd.read_csv('WineQT1.csv')
df
```

```python
for i in range(len(df)):
  if df['quality'][i] <= 6:
    df.loc[i, 'quality'] = 0
  else:
    df.loc[i, 'quality'] = 1

for column in df.columns:
  if column == 'quality':
    continue
  else:
    df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())

df
```
